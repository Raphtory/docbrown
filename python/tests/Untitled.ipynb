```python
from raphtory import Graph
from raphtory import algorithms
from raphtory import Perspective
import tempfile
import csv
from hplotlib import mpl_draw
```

```python
# !curl -o /tmp/lotr.csv https://raw.githubusercontent.com/Raphtory/Data/main/lotr.csv

filename = '/tmp/lotr.csv'

graph = Graph(4)
with open(filename, 'r') as csvfile:
    datareader = csv.reader(csvfile)
    for row in datareader:

        source_node = row[0]
        destination_node = row[1]
        timestamp = int(row[2])
        
#         graph.add_vertex(timestamp, source_node, {})
#         graph.add_vertex(timestamp, destination_node, {})  
        graph.add_edge(timestamp, source_node, destination_node, {})
```

```python
graph.num_vertices(), graph.num_edges()
```

```python
mpl_draw(graph, with_labels=False)
```
